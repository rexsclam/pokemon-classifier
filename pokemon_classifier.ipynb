```python
from fastbook import *
from fastai.vision.widgets import *
import pandas as pd
```

```python
path = Path()
learn_inf = load_learner(path/'pokemon_cnn.pkl')
```

```python
btn_upload = widgets.FileUpload(layout=Layout(width='256px'))
```

```python
btn_run = widgets.Button(description='Who\'s That Pokemon?', layout=Layout(width='256px'))
```

```python
out_pl = widgets.Output()
```

```python
lbl_pred = widgets.Label()
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)

    classes = learn_inf.dls.vocab[:]
    probs_list = probs.tolist()
    class_to_result = list(zip(classes, probs_list))
    probs_df = pd.DataFrame(class_to_result, columns =['pokemon_name', 'probability']) 
    top_probs = probs_df.nlargest(5, 'probability')

    lbl_pred.value = top_probs.to_string(index=False, header=False, formatters={'probability': '{:,.2%}'.format})

btn_run.on_click(on_click_classify)
```

```python
#hide_output
VBox([widgets.Label('Upload your Pokemon!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
